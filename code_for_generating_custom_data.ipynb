```python
!pip install faker
from faker import Faker
import pandas as pd

fake = Faker()
#df = pd.read_csv('customer_support_tickets.csv').head(20)
df = pd.read_csv('/content/sample_data/customer_support_tickets.csv').head(20)

# Just add 3 columns with PII
df['Real_Email'] = [fake.email() for _ in range(20)]
df['Phone'] = [fake.phone_number() for _ in range(20)]
df['SSN'] = [fake.ssn() for _ in range(20)]

# Save it
df.to_csv('tickets_with_pii.csv', index=False)
```

```python
for i in range(5):
    row = df.iloc[i]
    with open(f'ticket_{i+1}.txt', 'w') as f:
        f.write(f"""
Ticket: {row['Ticket ID']}
Customer: {row['Customer Name']}
Email: {row['Real_Email']}
Phone: {row['Phone']}
SSN: {row['SSN']}

Issue: {row['Ticket Description']}
        """)
```
